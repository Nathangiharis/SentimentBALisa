**5.3 Tokenizing**

Tokenizing merupakan metode untuk membuat kalimat menjadi data perkata


```python
import nltk
nltk.download('punkt')
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
import re
from nltk.tokenize import word_tokenize


# Creating a list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to the list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('brand ambassador ajaib').get_items()):
    if i > 300:
        break
    raw_text = re.sub(r'@\w+|#\w+|\bhttps?://\S+\b|\d+|[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F1E0-\U0001F1FF]+', ' ', tweet.content)  # Clean text by removing mentions, hashtags, links, numbers, and emoticons
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', raw_text)  # Keep only Bahasa (Indonesian) language text
    tokens = word_tokenize(cleaned_text.strip().lower())  # Tokenize the text
    tweets_list1.append([tweet.date, tweet.id, tokens, tweet.user.username])  # Convert to lowercase and add tokens

# Creating a DataFrame from the tweets list above
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Tokens', 'Username'])

# Export scraped data to a CSV file
tweets_df1.to_csv('DataUAS_Tokenization.csv', index=False)
```

**5.4 Klasifikasi**

Melakukan klasifikasi berdasarkan data negatif, positif, dan neutral.

```python
import nltk
nltk.download('punkt')
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
import re
from nltk.tokenize import word_tokenize


# Creating a list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to the list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('brand ambassador ajaib').get_items()):
    if i > 300:
        break
    raw_text = re.sub(r'@\w+|#\w+|\bhttps?://\S+\b|\d+|[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F1E0-\U0001F1FF]+', ' ', tweet.content)  # Clean text by removing mentions, hashtags, links, numbers, and emoticons
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', raw_text)  # Keep only Bahasa (Indonesian) language text
    tokens = word_tokenize(cleaned_text.strip().lower())  # Tokenize the text
    tweets_list1.append([tweet.date, tweet.id, tokens, tweet.user.username])  # Convert to lowercase and add tokens

# Creating a DataFrame from the tweets list above
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'Tweet Id', 'Tokens', 'Username'])

# Define the positive and negative word lists
positive_words = ['happy', 'bangga', 'prestasi', 'congratulations', 'selamat', 'excited', 'proud', 'keren', 'welcoming']
negative_words = ['negatif', 'threat', 'kurang', 'jealous', 'gila', 'bakar uang', 'bakar duit', 'kurang cocok', 'mahal', 'aneh', 'tidak','down']

# Create a function to classify the sentiment based on positive and negative words
def classify_sentiment(tokens):
    positive_count = sum(token in positive_words for token in tokens)
    negative_count = sum(token in negative_words for token in tokens)

    if positive_count > negative_count:
        return 'Positive'
    elif positive_count < negative_count:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the sentiment classification function to each tweet in the DataFrame
tweets_df1['Sentiment'] = tweets_df1['Tokens'].apply(lambda tokens: classify_sentiment(tokens))

# Export scraped data to a CSV file
tweets_df1.to_csv('DataUAS_SentimentClassification.csv', index=False)
```

**5.5 Analisis dan Hasil**

Data yang terpakai sebanyak 301 data, kemudian data ini dibagi menjadi 2 bagian untuk data latih sebanyak 181 data. Dengan menggunakan data latih pada metode naÃ¯ve bayes didapat hasil akurasi sebesar 0.934 atau 93.4%.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tweets_df1 = pd.read_csv('DataUAS_SentimentClassification.csv')

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    tweets_df1['Tokens'], tweets_df1['Sentiment'], test_size=0.2, random_state=42
)

# Create a Bag-of-Words representation of the text data
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_train_bow = vectorizer.fit_transform(X_train)
X_test_bow = vectorizer.transform(X_test)

# Train the Naive Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_bow, y_train)

# Predict the sentiment labels for the testing data
y_pred = nb_classifier.predict(X_test_bow)

# Compute the confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)

# Calculate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)

# Visualize the confusion matrix
sns.heatmap(confusion_mat, annot=True, cmap="Blues", fmt="d", xticklabels=["Negative", "Neutral", "Positive"], yticklabels=["Negative", "Neutral", "Positive"])
plt.xlabel("Predicted Sentiment")
plt.ylabel("True Sentiment")
plt.title("Confusion Matrix")
plt.show()

# Print the accuracy
print("Accuracy:", accuracy)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tweets_df1 = pd.read_csv('DataUAS_SentimentClassification.csv')

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    tweets_df1['Tokens'], tweets_df1['Sentiment'], test_size=0.2, random_state=42
)

# Create a Bag-of-Words representation of the text data
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_train_bow = vectorizer.fit_transform(X_train)
X_test_bow = vectorizer.transform(X_test)

# Train the Naive Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_bow, y_train)

# Predict the sentiment labels for the testing data
y_pred = nb_classifier.predict(X_test_bow)

# Calculate precision, recall, and F1-score
report = classification_report(y_test, y_pred)

# Print the classification report
print(report)

# Convert the classification report to a pandas DataFrame
report_df = pd.DataFrame(classification_report(y_test, y_pred, output_dict=True)).T

# Display the precision, recall, and F1-score as a table
print(report_df)

# Visualize the precision, recall, and F1-score using a bar chart
metrics = ['precision', 'recall', 'f1-score']
scores = report_df.loc['macro avg', metrics]

sns.set(style='whitegrid')
plt.figure(figsize=(8, 5))
sns.barplot(x=metrics, y=scores, palette='muted')
plt.title('Precision, Recall, and F1-Score')
plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.show()
```
